## 总结

1.评论数前10、前5的智能产品分别占总评论数的96.6%、83.7%，以评论数作为销量的衡量标准，销量最高的智能产品是B018Y229OU（Fire TV）,其次是B00L9EPT8O,B01E6AO69U（Echo (White)）、B00U3FPN4U（Kindle Paperwhite）,B00OQVZDJM（All-New Fire HD 8 Tablet）,B01AHB9CN2（Amazon Fire Tv）。

2.亚马逊智能产品月均评论数是797.6条。评论数整体呈上涨趋势，且存在爆发式增长现象，在2016年1月、2017年1月评论数分别达到阶段性顶峰，2017年1月评论数较前一次明显增加，说明智能产品的潜在用户较多，市场前景好。

3.亚马逊智能产品平均评分4.59（±0.72），中位数为5，好评率93.5%，说明评分较高且集中，客户整体满意度较高。主要产品中，在评论量前5的产品好评率均在整体水平之上，其中B018Y229OU（Fire TV）好评率、推荐率最低。

4.好评关键词包括：享受、操作、工作、手持、价格；差评关键词包括：费用、质量、老旧、屏幕、问题、连接等。

## 建议

1.月度评论数趋势阶段性上涨趋势明显，每年1月份前后是智能产品销售的旺季，可在在新品研发、备货、广告文案、渠道推广、客户服务等方面提前布局。结合产品评论数，对产品评论数前10的产品进行重点推广。
